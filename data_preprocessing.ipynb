```python
# DATA PREPROCESSING
import numpy as np
import pandas as pd
data=pd.read_csv("/content/Restaurant reviews (1).csv")
data.head()
```

```python
shape = data.shape

print(f'The dataset has {shape[0]} rows and {shape[1]} columns.')
```

```python
data.info()
```

```python
data.describe()
```

```python
#Check for missing values
print(data.isnull().sum())
```

```python
# Fill missing values in 'Reviewer', 'Review', 'Rating', 'Metadata', and 'Time' columns with a placeholder
placeholder = 'Unknown'
data['Reviewer'].fillna(placeholder, inplace=True)
data['Review'].fillna(placeholder, inplace=True)
data['Rating'].fillna(placeholder, inplace=True)
data['Metadata'].fillna(placeholder, inplace=True)
data['Time'].fillna(placeholder, inplace=True)
```

```python
# Fill missing values in 'Reviewer', 'Review', 'Rating', 'Metadata', and 'Time' columns
import pandas as pd
from tqdm import tqdm

tqdm.pandas()



# Impute missing values
text_columns = ['Restaurant', 'Reviewer', 'Review', 'Metadata', 'Time']
numeric_columns = ['Rating', 'Pictures']

data[text_columns] = data[text_columns].fillna('Missing')
data[numeric_columns] = data[numeric_columns].fillna(-1)

# Display the head of the dataframe to confirm changes
print(data.head())
```

```python
# Convert 'Rating' column to numeric, coerce errors to  and then fill  with a placeholder value
# Assuming ratings are on a scale of 1-5
#errors='coerce': This parameter specifies that any values that cannot be converted to numeric types should be set to NaN (missing values).
#This is useful for handling cases where the 'Rating' column may contain non-numeric values or strings that cannot be converted to numbers.
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce').fillna(-1).astype(int)
print(data['Rating'])
```

```python
print(data['Time'])
```

```python
# Convert 'Time' column to datetime
data['Time'] = pd.to_datetime(data['Time'], errors='coerce')
print(data['Time'])
```

```python
# Remove duplicate rows
initial_row_count = data.shape[0]
data.drop_duplicates(inplace=True)
final_row_count = data.shape[0]
print(initial_row_count)
print(final_row_count )
```

```python
# Standardize text data by trimming whitespace and converting to lowercase
text_columns = ['Restaurant', 'Reviewer', 'Review', 'Metadata']
for col in text_columns:
    data[col] = data[col].str.strip().str.lower()

print(data[col])
data.head()
```

```python
print('Duplicates removed and text data standardized.')
print('Rows removed:', initial_row_count - final_row_count)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Save the training and testing sets to CSV files
train_data.to_csv('train_reviews.csv', index=False)
test_data.to_csv('test_reviews.csv', index=False)
print(train_data)
```
